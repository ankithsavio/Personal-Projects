```python
data = []
for text in df['Data']:
    data1 = re.sub("\s+"," ",text).lower() 
    data2 = re.sub("\[.+\]","",data1)
    data3 = re.sub("&.+;","",data2)
    data4 = re.sub(r'http\S+', '',data3)
    data.append(re.sub(r'[^\w\s]',"",data4))
```

```python
df['Data'] = data
```

```python
stop_words=set(stopwords.words("english"))
wnet = WordNetLemmatizer()
port = PorterStemmer()
stop_words_2 = []
condition = ['not','nor','no']
for words in stop_words:
    if words not in condition:
        stop_words_2.append(words)

def lemmatize_text(text):
    words = word_tokenize(text)
    words_2 = []
    lemm_2 = ""
    for word in words:
        if word not in stop_words_2:
            words_2.append(word)
    for word in words_2:
        lemm = wnet.lemmatize(word)
        lemm_2+=lemm+" "
    return lemm_2
```

```python
df['Lemmatized Text'] = df['Data'].apply(lambda x:lemmatize_text(x))
```

```python
df
```

```python
lemm_desc = ""
for text in df['Lemmatized Text']:
    lemm_desc+=text+" "
```

```python
language = "en"
max_ngram_size = 2
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 500

custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(lemm_desc)

for kw in keywords:
    print(kw)
```

```python
def sentiment_score(text):
    score = SentimentIntensityAnalyzer().polarity_scores(text)
    if score['pos']>0:
        return 1
    else:
        return 0
```

```python
score = []
for text in df['Lemmatized Text']:
    score2 = sentiment_score(text)
    score.append(score2)
```

```python
df["Sentiment score"] = score
```
